```python
fig = px.scatter(title='Embalses Asturias comparación en porcentaje')

fig.add_scatter(x=df1.index, y=df1['Alfilorios Porcentaje'], mode='markers', name='Alfilorios')
fig.add_scatter(x=df1.index, y=df1['Arbón Porcentaje'], mode='markers', name='Arbón')
fig.add_scatter(x=df1.index, y=df1['La Barca Porcentaje'], mode='markers', name='La Barca')
fig.add_scatter(x=df1.index, y=df1['Doiras Porcentaje'], mode='markers', name='Doiras')
fig.add_scatter(x=df1.index, y=df1['Salime Porcentaje'], mode='markers', name='Salime')
fig.add_scatter(x=df1.index, y=df1['Tanes - Rioseco Porcentaje'], mode='markers', name='Tanes - Rioseco')
fig.update_layout(
    title="Comparación embalses Asturias en porcentaje",
    xaxis_title="Tiempo",
    yaxis_title="Porcentaje de llenado(%)",
    legend_title="Embalses",
    font=dict(
        family="Courier New, monospace",
        size=18,
        color="Gray"
    )
)
fig.show()
```

```python
fig = px.scatter()

fig.add_scatter(x=df1.index, y=df1['Alfilorios Volumen (hm3)'], mode='markers', name='Alfilorios')
fig.add_scatter(x=df1.index, y=df1['Arbón Volumen (hm3)'], mode='markers', name='Arbón')
fig.add_scatter(x=df1.index, y=df1['La Barca Volumen (hm3)'], mode='markers', name='La Barca')
fig.add_scatter(x=df1.index, y=df1['Doiras Volumen (hm3)'], mode='markers', name='Doiras')
fig.add_scatter(x=df1.index, y=df1['Salime Volumen (hm3)'], mode='markers', name='Salime')
fig.add_scatter(x=df1.index, y=df1['Tanes - Rioseco Volumen (hm3)'], mode='markers', name='Tanes - Rioseco')
fig.update_layout(
    title="Comparación embalses Asturias en capacidad (hm3)",
    xaxis_title="Tiempo",
    yaxis_title="Volumen (hm3)",
    legend_title="Embalses",
    font=dict(
        family="Courier New, monospace",
        size=18,
        color="Gray"
    )
)
fig.show()
```

```python
df1['Volumen total (hm3)'] = df1['Alfilorios Volumen (hm3)'] + df1['Arbón Volumen (hm3)'] + df1['Doiras Volumen (hm3)'] + df1['Salime Volumen (hm3)'] + df1['Tanes - Rioseco Volumen (hm3)'] + df1['La Barca Volumen (hm3)']
df1['Porcentaje total'] = ((df1['Alfilorios Porcentaje'] + df1['Arbón Porcentaje'] + df1['Doiras Porcentaje'] + df1['Salime Porcentaje'] + df1['Tanes - Rioseco Porcentaje'] +df1['La Barca Porcentaje']) /6).round(2)
df1
```

```python
df3 = df2[df2.index>116]
df3 = df3.copy()
print(df3.columns.tolist())
df3['Periodo'] = pd.to_datetime(df3['Periodo'], format='%Y-%m')
df3.set_index('Periodo', inplace=True)
```

```python
df3
```

```python
ambos = pd.concat([df1['Volumen total (hm3)'], df3['Aeropuerto Precipitación total (mm)'], df1['Porcentaje total']], axis=1)
ambos = ambos.copy()
ambos['Aeropuerto Precipitación total (mm)'] = pd.to_numeric(ambos['Aeropuerto Precipitación total (mm)'], errors='coerce')
print(ambos.dtypes)
# ambos['Aeropuerto Precipitación total (mm)'] = ambos['Aeropuerto Precipitación total (mm)']

# ambos['Aeropuerto Precipitación total (mm)'] = pd.to_numeric(ambos['Aeropuerto Precipitación total (mm)'])
ambos['Precipitacion escalada'] = (ambos['Aeropuerto Precipitación total (mm)']/ambos['Aeropuerto Precipitación total (mm)'].max() * 100).round(2)
ambos
```

```python
fig = px.bar(ambos, x=ambos.index, y=ambos['Volumen total (hm3)'])

fig.add_scatter(x=ambos.index, y=ambos['Aeropuerto Precipitación total (mm)'], mode='lines+markers', name='Precipitaciones (mm)')
```

```python
fig = px.bar(ambos, x=ambos.index, y=ambos['Porcentaje total'])

fig.add_scatter(x=ambos.index, y=ambos['Precipitacion escalada'], mode='lines+markers', name='Precipitaciones escalada (mm)')
```
